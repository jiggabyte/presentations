# Prompt Engineering.     

#### Prompting AI Models is a novel field that encapsulates the art of mining information from an AI model, using informed and certain techniques to prune valuable and relavant responses from the AI models.

### Getting data from Gen AI models really takes an art form, as the instructions giving to the AI app need to be well crafted to get better responses. This has giving birth to a new ever changing and dynamic field referred to as "Prompt Engineering" as careful strategies are employed to get the better responses from Gen AI.

## Techniques in Prompting Gen AI     


There are about 3 types of techniques used in prompting Gen AI for better responses.
- The Instruction technique
- Role prompting
- Few Shot Prompting
- Combination Technique

### The Instruction technique uses the pattern of giving direct instructions to the AI chatbot to process the instructions and return a response.


### The Role Prompting technique prives a role description tothe Gen AI to assume and use to process and response to the question.

###  Few Shot Prompting provides certain examples to the Gen AI to use as seed to process and provide response.

### Combination technique uses multiple patterns, combining few shot and instructions or using instructions and role prompting to process and respond to a question or in no particular other using mixtures of different techniques to process a question and return a response.
