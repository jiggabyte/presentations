# JorDoctorAI

 




#### Your personal AI medical advicer.

JordoctorAI provides medical advice to users after considering symptoms and details giving to it while recommending that users seek a physical medical doctor for serious and emmergency cases. Users will launch the app and describe certain symptoms or conditions and the app will respond using AI. The app will also have a medical directory where medical centres or hospitals will be gleaned from google maps processed and return to the users as the search for hospitals around them or in area covered by google maps.

## Architecture

#### The app will utilise Gradio for UI and Python for Back End using ChatGPT API for AI Intelligence
